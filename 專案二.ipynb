```python
def clean(data):
    data = data.drop(["PassengerId", "Name", "Cabin"], axis=1)

    cols = ["VRDeck","ShoppingMall","RoomService","Spa","FoodCourt","Age"]#都是數值資料
    for col in cols:
        data[col].fillna(data[col].mean(), inplace=True) #以平均值補缺值，inplace=true: 把平均值assign給那些空格

    return data
```

```python
train = clean(train)
test = clean(test)
```

```python
train_x = pd.DataFrame(train.iloc[:,:-1])
train_y = pd.DataFrame(train.iloc[:,-1])
print(f'the shape of train_x is {train_x.shape}')
print(f'the shape of test_y is {test_y.shape}')
test_x = test.iloc[:,:-1]
test_y = test.iloc[:,-1]
```

```python
train_y=pd.get_dummies(train_y)
train_y.shape
```

```python
from operator import concat
concat_data = pd.concat([train_x, test],axis = 0)
concat_dummy = pd.get_dummies(concat_data)
dummy_train = concat_dummy.iloc[:8693,:]
dummy_test = concat_dummy.iloc[8693:,:]
print(f'train data size:{dummy_train.shape}')
```

```python
dummy_train.head(10)
```

```python
from sklearn.model_selection import train_test_split
train_x, test_x, train_y, test_y  = train_test_split(dummy_train, train_y, test_size=0.2, random_state=10)

from sklearn.preprocessing import StandardScaler
scaler_x = StandardScaler()
sc_train_x = scaler_x.fit_transform(train_x)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(sc_train_x,train_y)
```

```python
from sklearn.metrics import confusion_matrix
sc_test_x = scaler_x.transform(dummy_test)
y_hat = knn.predict(sc_test_x)
print(y_hat)
submi = pd.DataFrame({'PassengerID':id,'Transported':y_hat})
```
